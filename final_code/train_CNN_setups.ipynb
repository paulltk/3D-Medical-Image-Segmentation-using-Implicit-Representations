### Golden CNN

```python
evaluations = [
            #    ("reconstruction", None), 
               ("segmentation_bin", "binary"), 
               ("segmentation_sdf", "sdf")
               ]

for evaluation, seg_type in evaluations: 
    for i in range(3):
        ARGS = init_ARGS()
    
        # Setup changes here
        setup = "golden_cnn"
        ARGS.cnn_setup = "golden"

        ARGS.name = f"{evaluation}_{setup}_run_{i}"
        print(f"Starting training {ARGS.name}.")
        if "reconstruction" in evaluation: 
            print(vars(ARGS))
            
            reconstruction(ARGS) 
        
        if "segmentation" in evaluation: 
            ARGS.segmentation = seg_type
            print(vars(ARGS))
            
            segmentation(ARGS) 
        
        torch.cuda.empty_cache()    
```
