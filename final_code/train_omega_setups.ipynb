### Golden CNN

```python
evaluations = [
               ("reconstruction", None), 
               ("segmentation_bin", "binary"), 
               ("segmentation_sdf", "sdf")
               ]

omega_0s = [100, 10, 300, 30]

for omega_0 in omega_0s:
    for evaluation, seg_type in evaluations: 
        for i in range(1):
            ARGS = init_ARGS()
        
            # Setup changes here
            setup = "omega_" + str(omega_0) 
            ARGS.first_omega_0 = omega_0
            ARGS.pcmra_first_omega_0 = omega_0

            ARGS.name = f"{evaluation}_{setup}_run_{i}"
            print(f"Starting training {ARGS.name}.")
            if "reconstruction" in evaluation: 
                print(vars(ARGS))
                
                reconstruction(ARGS) 
            
            if "segmentation" in evaluation: 
                ARGS.segmentation = seg_type
                print(vars(ARGS))
                
                segmentation(ARGS) 
            
            torch.cuda.empty_cache()    
```
