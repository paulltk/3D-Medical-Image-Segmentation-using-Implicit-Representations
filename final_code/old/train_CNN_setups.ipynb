### Max Pooling CNN

```python
training_setups = [
               ("reconstruction", None, "maxpool", "golden"), 
               ("segmentation", "binary", "maxpool", "golden"), 
               ("segmentation", "sdf", "maxpool", "golden"),
               
               ("reconstruction", None, "batchnorm", "golden"), 
               ("segmentation", "binary", "batchnorm", "golden"), 
               ("segmentation", "sdf", "batchnorm", "golden"),
               
               ("reconstruction", None, "deep", "golden_deep"), 
               ("segmentation", "binary", "deep", "golden_deep"), 
               ("segmentation", "sdf", "deep", "golden_deep"),

               ("reconstruction", None, "golden", "golden"), 
               ("segmentation", "binary", "golden", "golden"), 
               ("segmentation", "sdf", "golden", "golden"),
               ]

for i in range(1):
    for training_setup, segmentation, cnn_setup, mapping_setup in training_setups: 
        ARGS = init_ARGS()
        
        ARGS.training_setup = training_setup
        ARGS.segmentation = segmentation

        ARGS.cnn_setup = cnn_setup
        ARGS.mapping_setup = mapping_setup

        print(f"Starting training {ARGS.name}.")

        print(vars(ARGS))
            
        complete_training(ARGS)  
                
        torch.cuda.empty_cache()      
```
