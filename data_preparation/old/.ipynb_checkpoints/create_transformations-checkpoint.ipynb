```python
axes = [(0, 1), (0, 2), (1, 2)]
angles = [-6, -4, -2, 2, 4, 6]
axes_angles = list(it.product(axes, angles))

os.listdir(root)

for file in os.listdir(root): 
    print(file)
    
    if file == "subjects.npy":
        continue

    subj, proj, trans, imtype = file.split("__")
        
    image = np.load(os.path.join(root, file))

    random.shuffle(axes_angles)
    subj_axes_angles = axes_angles
    
    for axis, angle in subj_axes_angles: 
        new_file = f"{subj}__{proj}__r {angle} {axis}__{imtype}"

        new_image = scipy.ndimage.interpolation.rotate(image.astype(np.float64), angle, axis, reshape=False)
        np.save(root + new_file, new_image.astype(np.float16))
```
