## Aorta Volunteers

```python
project = Project("Aorta Volunteers")
for i in range(project.get_number_of_subjects()):
    subject, pcmra, mask = project.get_subject(i) 
    pcmra = pcmra.flatten()
    
    plt.hist(pcmra, bins=100)
    
    plt.gca().set(title=str(i) + "   " + subject, ylabel='Frequency');
    
    plt.show()

    print("97% \t", np.percentile(pcmra, 97))
    print("98% \t", np.percentile(pcmra, 98))
    print("99% \t", np.percentile(pcmra, 99))
    print("99.5% \t", np.percentile(pcmra, 99.5))
    print("100% \t", np.percentile(pcmra, 100))
    
    
```
