```python
import numpy as np 
import os 
import matplotlib.pyplot as plt
from scipy.ndimage import zoom
```

```python
root = "/home/ptenkaate/scratch/Data/"
projects =  ["Aorta Volunteers", "Aorta BaV", "Aorta Resvcue", "Aorta CoA"]

save_to = "/home/ptenkaate/scratch/Master-Thesis/Dataset"

correct_shape = (128, 128, 24)

scale = (0.5, 0.5, 1)

for project in projects: 
    print(project)
    p_path = os.path.join(root, project, "pcmra")
    m_path = os.path.join(root, project, "mask")
    
    for subject in os.listdir(p_path): 
        pcmra = np.load(os.path.join(p_path, subject))
        try: 
            mask = np.load(os.path.join(m_path, subject))
        except: 
            continue
        
        if pcmra.shape != mask.shape:
            continue

        subject = subject[:-4]
        print(subject)
        
        pcmra = pcmra / pcmra.max()
        mask = mask / mask.max()
        
        r_string = ""
        if pcmra.shape != correct_shape: 
            resize = tuple([i/ j for i, j in zip(correct_shape, pcmra.shape)])
            
            pcmra = zoom(pcmra, resize)
            mask = zoom(mask, resize).round()
            
            resize = tuple([round(i, 1) for i in list(resize)])
            r_string = " resized " + str(resize) 
        
        p_name = f"{subject}__{project}__original{r_string}__pcmra.npy"
        m_name = f"{subject}__{project}__original{r_string}__mask.npy"
        
        np.save(os.path.join(save_to, "original_normalized", p_name), pcmra.astype(np.float16))
        np.save(os.path.join(save_to, "original_normalized", m_name), mask.astype(np.float16))
        
        #### save scaled images ####
        small_pcmra = zoom(pcmra, scale)
        small_mask = np.round(zoom(mask, scale))
        
        small_p_name = f"{subject}__{project}__scaled{r_string}__pcmra.npy"
        small_m_name = f"{subject}__{project}__scaled{r_string}__mask.npy"
            
        np.save(os.path.join(save_to, "scaled_normalized", small_p_name), small_pcmra.astype(np.float16))
        np.save(os.path.join(save_to, "scaled_normalized", small_m_name), small_mask.astype(np.float16))
        
        
```
