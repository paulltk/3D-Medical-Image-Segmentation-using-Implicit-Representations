```python
import scipy.io
import numpy as np
import os
from scipy.ndimage import zoom
```

```python
root = "/home/ptenkaate/lood_storage/divi/Projects/4dflow/PimWork/MachineLearningSegmentationMartijn/data_for_ML/"
folders = ["data_c1", "data_c2", "data_p1", "data_p2", "data_p3"]

save_loc = "/scratch/ptenkaate/Master-Thesis/Dataset/new_original"
```

```python
correct_shape = (128, 128, 24) 

for folder in folders: 
    path = os.path.join(root, folder)
    for file in os.listdir(path): 
        if "pc_mra" in file or "mask" in file: 
            print(file)

            file_path = os.path.join(path, file)
            # load numpy array
            mat = scipy.io.loadmat(file_path) 
            
#             print(mat.keys())
            
            if "pc_mra" in file: 
                image = mat["pc_mra"]
            else:
                image = mat["mask"]
                
            
            # reshape image 
            r_string = ""
            if image.shape != correct_shape: 
                print("reshaping")
                resize = tuple([i/ j for i, j in zip(correct_shape, image.shape)])
    
                image = zoom(image, resize)
                   
                r_string = " resized " + str(tuple([round(i, 1) for i in list(resize)])) 
            
                if "mask" in file: 
                    image = image.round()
                    
            # normalize image 
            if "pc_mra" in file:
                mx = np.percentile(image[:, 30:-30, :], 99.8)
                mn = np.percentile(image[:, 30:-30, :], 0.02)
            
                image = (image - mn) / (mx - mn)
                
                image = np.clip(image, 0, 1)
            
            if "pc_mra" in file: 
                imtype = "pcmra"
            elif "mask" in file: 
                imtype = "mask"
            else: 
                raise(Exception("blah"))
            
            #create name to be saved
            name  = f"{file.split('.')[0].split('_')[0]}__{folder}__original{r_string}__{imtype}.npy"
            
            
            save_path = os.path.join(save_loc, name)
            np.save(save_path, image)
```
